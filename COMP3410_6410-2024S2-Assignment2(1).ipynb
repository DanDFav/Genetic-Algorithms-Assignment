```python
def add_times(time1, time2):
    # Convert both times to minutes and seconds
    minutes1, seconds1 = divmod(time1 * 100, 100)
    minutes2, seconds2 = divmod(time2 * 100, 100)
    
    # Get the total minutes and total seconds
    total_minutes = minutes1 + minutes2
    total_seconds = seconds1 + seconds2
    
    # If the amount of seconds is greater than 60, we increment minutes, 
    # and decrement seconds
    if total_seconds >= 60:
        total_minutes += 1
        total_seconds -= 60
    
    # We get the actual time in minutes and seconds 
    real_time = total_minutes + (total_seconds / 100)

    return round(real_time,2)
```

```python
# Same as addition, but we just subtract the times
def subtract_times(time1, time2):
    minutes1, seconds1 = divmod(time1 * 100, 100)
    minutes2, seconds2 = divmod(time2 * 100, 100)
    
    total_minutes = minutes1 - minutes2
    total_seconds = seconds1 - seconds2
    
    if total_seconds < 0:
        total_minutes -= 1
        total_seconds += 60
    
    real_time = total_minutes + (total_seconds / 100)
    
    return round(real_time, 2)
```

```python
# provide your code here
# around 7 simple functions are needed to complete this task. 
# generate_initial_population - Done
# fitness_function - Partial 
# selection - Kinda Done
# crossover - Unsure
# exchange 
# successor
# genetic_algorithm

import copy
import math

def generate_initial_population(): 
    global time
    
    population = {}

    # Simulate 500 ant colony simulations, and get the result of each one
    for i in range(500): 
        population[i] = setup()
        print(i, ": Finished at: ", time)

    original = copy.deepcopy(population)
    
    return population, original


def fitness_function(old_generation): 
    survivors = []

    # Shuffle the population and get the midpoint
    random.shuffle(old_generation)
    middway = len(old_generation) // 2

    # Iterate over half the population, compare x and x + midpoint, 
    # add which ever has the lowest time / damage 
    for i in range(middway): 
        if old_generation[i]['damage'] < old_generation[i + middway]['damage']:
            survivors.append(old_generation[i])
        elif old_generation[i]['damage'] == old_generation[i + middway]['damage']:
            if old_generation[i]['time'] <= old_generation[i + middway]['time']:
                survivors.append(old_generation[i])
            else: 
                survivors.append(old_generation[i + middway])
        else: 
            survivors.append(old_generation[i + middway])

    return survivors

def choose_successors(generation): 
    exchange_group = []
    for path in generation: 
        # Only 10% of our individuals will be altered 
        if random.randint(0, 100) <= 20: 
            exchange_group.append(exchange(path))
        else: 
            exchange_group.append(path)
    return exchange_group

def crossover(population): 
    offspring = []
    n = len(population)

    # For each entry, create an offspring by crossing-over with another entry 
    for i in range(n): 
        parent_a = population[i]
        parent_b = population[(i + n // 2) % n]  
        offspring.append(create_offspring(parent_a, parent_b))
        offspring.append(create_offspring(parent_b, parent_a))
    return offspring


def create_offspring(a, b, num_keys=1): 
    parent_a = copy.deepcopy(a)
    parent_b = copy.deepcopy(b)

    # Get the path each unit took
    a_path = parent_a['path']
    b_path = parent_b['path']

    # Get a random unit from the simulation
    keys_to_swap = random.sample(list(a_path.keys()), min(num_keys, len(a_path)))

    # For however many keys there are, swap the path that unit took, 
    # from parent b to parent a
    for key in keys_to_swap:
        parent_a['path'][key] = copy.deepcopy(b_path[key])


    return parent_a

def exchange(individual): 
    paths = individual['path']
    
    # Swap 2 regions of all units in this simulation, 
    # as long as it is still valid  
    for agent, subpaths in paths.items():
        for path in subpaths:
            # If it is less than three, we cannot create a valid
            # sequence 
            if len(path) >= 3:
                c1_index, c2_index = random.sample(range(1, len(path)), 2)
                path[c1_index], path[c2_index] = path[c2_index], path[c1_index]


    return individual
```

```python
def is_2d_array(array):
    # Checks if the array is 2d or not 
    if isinstance(array, list) and len(array) > 0:
        return True
    return False
```
