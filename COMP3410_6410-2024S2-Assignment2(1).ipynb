The city is divided into **5 regions (R1 to R5)**. Each region has an evaluated **damage level** according to its blocked roads and fires: 
- High damage (H): $>=50\%$ roads blocked, or $>=30\%$ fires; 
- Medium damage (M): $[10\%, 50\%)$ roads blocked, or $[10\%, 30\%)$ fires; and 
- Low damage (L): $<10\%$ roads blocked, or $<10\%$ fires. 

Either blocked roads or fires that reach the range can be considered its damage level. For example, a region with 30\% road blocks and 5\% fires is Medium damaged. 

The **initial damage** to each region is provided: 
- R1 (H): 60\% roads blocked, 35\% fires; 
- R2 (M): 40\% roads blocked, 25\% fires;
- R3 (M): 15\% roads blocked, 5\% fires; 
- R4 (H): 35\% roads blocked, 30\% fires; and 
- R5 (L): 5\% roads blocked, 3\% fires.

Assume all regions are of equal size to simplify the modelling process. Each region is modelled as a **node** in a graph, and predefined **distances** between regions (e.g., 5 km between R1 and R2, 3 km between R2 and R3) are detailed: R1 -- R2: 5 km, R1 -- R3: 7 km, R1 -- R4: 4 km, R1 -- R5: 6 km, R2 -- R3: 3 km, R2 -- R4: 4 km, R2 -- R5: 8 km, R3 -- R4: 5 km, R3 -- R5: 6 km, and R4 -- R5: 4 km.

Aftershocks occur randomly every 15 minutes increasing road blockages by 10\% over all regions. For example, in R1, the initial damage is 60\% roads blocked and 35\% fires. After 10 minutes, it becomes $(60\%, 45\%)$, and after an additional 5 minutes, it becomes $(70\%, 45\%)$ if no rescue operations are performed.

Rescue agents are distributed across regions at the start of the simulation. **Eight units of fire trucks start at R2.** Each fire truck unit decreases fire percentage by 10\% per rescue operation. **Six units of police start at R4.** Each police unit decreases road blockages by 10\% per rescue operation. Multiple agents/units can perform rescue operations at the same time in the same region. The effects of their actions are cumulative. For example, if 2 fire trucks are deployed to R1 simultaneously, the fire percentage decreases by 20\% in one operation. Rescue operations affect the regional damage directly. If a fire truck reduces fire percentage by 10\%, that 10\% decrease is applied to the overall fire damage in that region, e.g., $20\%-10\%=10\%$. 

**Assumption:** Both rescue operations and disaster events (fire spread and aftershocks) do not consume time in this simulation, but each unit can only perform rescue once when visiting the region. After completing their assigned rescue tasks along the path, they must return to their starting point for a refill.

**Travel Speed:** Rescue agents can travel at a maximum speed of 60 km/h on unblocked roads.
```python
def add_times(time1, time2):
    # Convert both times to minutes and seconds
    minutes1, seconds1 = divmod(time1 * 100, 100)
    minutes2, seconds2 = divmod(time2 * 100, 100)
    
    # Get the total minutes and total seconds
    total_minutes = minutes1 + minutes2
    total_seconds = seconds1 + seconds2
    
    # If the amount of seconds is greater than 60, we increment minutes, 
    # and decrement seconds
    if total_seconds >= 60:
        total_minutes += 1
        total_seconds -= 60
    
    # We get the actual time in minutes and seconds 
    real_time = total_minutes + (total_seconds / 100)

    return round(real_time,2)
```

```python
# Same as addition, but we just subtract the times
def subtract_times(time1, time2):
    minutes1, seconds1 = divmod(time1 * 100, 100)
    minutes2, seconds2 = divmod(time2 * 100, 100)
    
    total_minutes = minutes1 - minutes2
    total_seconds = seconds1 - seconds2
    
    if total_seconds < 0:
        total_minutes -= 1
        total_seconds += 60
    
    real_time = total_minutes + (total_seconds / 100)
    
    return round(real_time, 2)
```

Similar to the calculation of the averaged blockages between two regions, the pheromone level can be determined by averaging the conditions of the two connected regions. For example, $\rho^{\text{fire}}_{R_1R_2} = 1 - [\text{fire damage}(R1)+\text{fire damage}(R2)]/2/100$.

Additionally, the equations are provided for your convenience in reporting.

$$\tau_{ij} \leftarrow \text{Evaporation}(\tau_{ij})$$ 

$$\tau_{ij} \leftarrow \text{Rescue}(\tau_{ij})$$
 
where `Evaporation` can be fire spread or aftershock, `Rescue` can be from a unit of fire truck or a unit of police.

$$\eta_{ij}=1/d_{ij}$$

$$p^k_{ij}=\frac{[\tau_{ij}]^\alpha[\eta_{ij}]^\beta}{\sum_{l\in N_k(i)}[\tau_{il}]^\alpha[\eta_{il}]^\beta}$$

```python
# provide your code here
# around 5 simple functions are needed to complete this task. 
# evaporate_pheromones
# deposit_pheromones
# choose_next_region
# edge_travel_time
# simulate_agent_path


import random 

def evaporate_pheromones(type):
    global region_damage
    if type == 'aftershocks': 
        for region in region_damage: 
            region_damage[region][0] = round(region_damage[region][0] + 0.10, 2)

    else: 
        for region in region_damage: 
            region_damage[region][1] = round(region_damage[region][1] + 0.10, 2)
    

def deposit_pheromones(agent, region): 
    global region_damage
    global firemen
    global police
    type = agent[0]

    # Reduce the damage by 10%, update the locations we've visited, add the location to our 
    # overall path
    if type == 'F': 
        region_damage[region][1] = round(max(0, region_damage[region][1] - 0.10), 2)
        firemen[agent]['visited_lcoations'].append(region)
        firemen[agent]['path'].append([region, time])
    else: 
        region_damage[region][0] = round(max(0, region_damage[region][0] - 0.10),2 )
        police[agent]['visited_lcoations'].append(region)
        police[agent]['path'].append([region, time])

def choose_next_region(agent, pheromone_list):
    global police 
    global firemen
    global time

    # Instantiate our variables
    visited_locations = None
    current_region = None
    refuel = None
    type = agent[0]

    # Get the regions we have been too, and our current region
    if type == 'P': 
        visited_locations = police[agent]['visited_lcoations']
        current_region = police[agent]['location']
        refuel = 'R4'
    else:
        visited_locations = firemen[agent]['visited_lcoations']
        current_region = firemen[agent]['location']
        refuel = 'R2'

    # If we have visited every region, return to the region we started at, 
    # and empty our visited regions array
    if len(visited_locations) == 5: 
        region = refuel
        if type == 'F': 
            firemen[agent]['visited_lcoations'] = []
        else: 
            police[agent]['visited_lcoations'] = []

    # If we still have regions to visit, choose a random region out of the 
    # available regions
    else:
        region_choice = [region for region in pheromone_list.keys() if region not in visited_locations]
        probability = [pheromone_list[region] for region in region_choice]

        # If everywhere has no pheromone levels, we choose completely at random
        if sum(probability) == 0: 
            region = random.choice(region_choice)
        
        # If there are pheromone levels, we choose based off the weight of those 
        # pheromones 
        else: 
            region = random.choices(region_choice, weights=probability, k=1)[0]
        print(agent, " Chose region: ", region)
        
    # Calculate the time it will take to travel to the region
    time_to = add_times(edge_travel_time(current_region, region), time)
    print(time_to)

    # Update our actions for each agent
    if type == 'F': 
        firemen[agent].setdefault('next_action', [region, time_to])
    else:
        police[agent].setdefault('next_action', [region, time_to]) 
    

def simulate_agent_path(): 
    global police 
    global firemen
    global time

    # While our time is less than 90 minutes, and we havent found a solution yet
    while time < 90.0 and not is_complete(): 

        # Every specified times we cause an aftershock and spread fire
        if time % 10.0 == 0.0 and time != 0.0: 
            evaporate_pheromones('spread_fire')
            
        if time % 15.0 == 0.0 and time != 0.0: 
            evaporate_pheromones('aftershocks')

        # For each agent, if they can act at this time, they act
        for agent in all: 
            if agent[0] == 'F' and firemen[agent].get('next_action'): 
                action_firemen(agent)
                    
            elif agent[0] == 'P' and police[agent].get('next_action'): 
                action_police(agent)
                
        time = add_times(time, 0.01)

def edge_travel_time(current_region, next_region):
    global travel_times
    global pheromone_matrix_blockage

    # Get the amount that the road blockages will slow the vehicle down
    slow_down = 1 - pheromone_matrix_blockage[current_region + next_region]
    new_speed = 60 * slow_down 

    # Get the amount of hours it will take to reach the region, then convert 
    # to seconds
    time_in_hours = travel_times[current_region + next_region] / new_speed
    time_in_seconds = time_in_hours * 3600 

    # Get the amount of minutes and seconds it will take. 
    minutes = int(time_in_seconds // 60)
    seconds = round(time_in_seconds % 60, 3)

    return round(minutes + (seconds / 100),2)

def get_pheromone_list(region, type): 
    global region_damage

    # A list of regions, excluding the region we are currently at 
    regions = [r for r in region_damage.keys() if r != region]
    
    # Creates a dictionary which holds the pheromone level from our region
    # to each other region
    pheromones = {}
    for i in regions: 
        pheromones[i] = type[i + region]
    
    total = sum(pheromones.values())
    
    # If our total is 0, we have visited everywhere, and need to refuel
    # we return a list of 0's
    if total == 0:
        normalized_pheromones = {key: 0 for key in pheromones.keys()}
    
    # Otherwise we return a dictionary, which holds the normalized values from 
    # our region to each other region
    else:
        normalized_pheromones = {key: round(value / total, 3) for key, value in pheromones.items()}

    return normalized_pheromones
```

```python
def generate_matricies(): 
    global travel_times
    global pheromone_matrix_fire
    global pheromone_matrix_blockage 
    
    # For link between cities, i.e R2 -> R3 = 3km, link = 'R2R3'
    for link in travel_times: 
        # Get the two regions 
        region_one = link[0:2]
        region_two = link[2:4]
        
        # Take the average of the fire or blockade between the two cites, that is the pheromone level for that path
        pheromone_matrix_fire[link] = round((region_damage[region_one][1] + region_damage[region_two][1]) / 2, 2)
        pheromone_matrix_blockage[link] = round((region_damage[region_one][0] + region_damage[region_two][0]) / 2, 2)
```

```python
def is_complete(): 
    global region_damage

    # Sum the values in our region_damage dict, if it's 0, we have finished the simulation
    total_block = sum(values[0] for values in region_damage.values())
    total_fire = sum(values[1] for values in region_damage.values())
    total = total_block + total_fire
    if total == 0: 
        return True
    else: 
        return False

def action_police(agent): 
    global police
    global region_damage

    agent_action = police[agent]['next_action'][0]
    agent_time = police[agent]['next_action'][1]
    agent_location = police[agent]['location']
    agent_time = subtract_times(agent_time, time)

    # If they have arrived at the location at this time
    if agent_time == 0.0:
        print(agent, ": ", agent_location, " | T = ", police[agent]['next_action'][1])
        
        # Reduce the blockades 
        deposit_pheromones(agent, agent_action)

        # Regenerate our new matricies with updated blockade levels 
        generate_matricies()

        # Update our location, and remove the 'next_action' field
        police[agent]['location'] = agent_action
        del police[agent]['next_action']

        # Check if we have finished 
        if is_complete(): 
            return
        print(region_damage)

        # Choose the next region to go to 
        choose_next_region(agent, get_pheromone_list(police[agent]['location'], pheromone_matrix_blockage))

def action_firemen(agent): 
    global firemen
    global region_damage
    
    agent_action = firemen[agent]['next_action'][0]
    agent_time = firemen[agent]['next_action'][1]
    agent_location = firemen[agent]['location']
    agent_time = subtract_times(agent_time, time)

    # If they have arrived at the location at this time
    if agent_time == 0.0:
        print(agent, ": ", agent_location, " | T = ", firemen[agent]['next_action'][1])
        
        # Reduce the fire 
        deposit_pheromones(agent, agent_action)
        
        # Regenerate our new matricies with updated fire levels 
        generate_matricies()

        # Update our location, and remove the 'next_action' field
        firemen[agent]['location'] = agent_action
        del firemen[agent]['next_action']

        # Check if we have finished 
        if is_complete(): 
            return
        print(region_damage)

        # Choose the next region to go to 
        choose_next_region(agent, get_pheromone_list(firemen[agent]['location'], pheromone_matrix_fire))
```
